#### 스타벅스 음료 크롤링

```python
import requests
import csv
from bs4 import BeautifulSoup
from selenium import webdriver

URL = "https://www.starbucks.co.kr/menu/drink_list.do"
star_menu = []

# 크롭 headless모드
chrome_option = webdriver.ChromeOptions()
chrome_option.add_argument("headless")
chrome_option.add_argument("-disable-gpu")
chrome_option.add_argument("lang-=ko_KR")

# 크룸 실행
dr = webdriver.Chrome("./chromedriver", chrome_options=chrome_option)

# URL 소스 가져오기
dr.get(URL)
html_code = dr.page_source

# 분석하기 쉽게 bs4로 변환
soup = BeautifulSoup(html_code, "html.parser")

# 매뉴 가져오기
def getMenuInfo(menu):
    name = menu.find("img")["alt"]
    img_Url = menu.find("img")["src"]

    return {"name": name, "img_Url": img_Url}

# csv로 변환
def csv_save(star_menu):
    f = open("beverage.csv", "w", encoding="utf-8", newline="")
    wr = csv.writer(f)
    wr.writerow(["이름", "URL"])
    for Info in star_menu:
        wr.writerow([Info["name"], Info["img_Url"]])
    f.close()


def main():
    menu_lists = soup.find_all("a", {"class": "goDrinkView"})
    for menu in menu_lists:
        star_menu.append(getMenuInfo(menu))

    csv_save(star_menu)


if __name__ == "__main__":
    main()
```

#### 스타벅스 푸드 크롤링

```python
import requests
import csv
from bs4 import BeautifulSoup
from selenium import webdriver

URL = "https://www.starbucks.co.kr/menu/food_list.do"
star_menu = []

# 크롭 headless모드
chrome_option = webdriver.ChromeOptions()
chrome_option.add_argument("headless")
chrome_option.add_argument("-disable-gpu")
chrome_option.add_argument("lang-=ko_KR")

# 크룸 실행
dr = webdriver.Chrome("./chromedriver", chrome_options=chrome_option)

# URL 소스 가져오기
dr.get(URL)
html_code = dr.page_source

# 분석하기 쉽게 bs4로 변환
soup = BeautifulSoup(html_code, "html.parser")

# 매뉴 가져오기
def getMenuInfo(menu):
    name = menu.find("img")["alt"]
    img_Url = menu.find("img")["src"]

    return {"name": name, "img_Url": img_Url}

# csv로 변환
def csv_save(star_menu):
    f = open("food.csv", "w", encoding="utf-8", newline="")
    wr = csv.writer(f)
    wr.writerow(["이름", "URL"])
    for Info in star_menu:
        wr.writerow([Info["name"], Info["img_Url"]])
    f.close()


def main():
    menu_lists = soup.find_all("a", {"class": "goFoodView"})
    for menu in menu_lists:
        star_menu.append(getMenuInfo(menu))

    csv_save(star_menu)


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
beverage_df = pd.read_csv('beverage.csv')
food_df = pd.read_csv('food.csv')
```

```python
beverage = beverage_df['이름'].to_list()
quantity = ['한','하나','둘','두','셋','세','네','석','다섯','여섯','일곱','여덟']
size = ['Short','Tall','Grande','Venti']
syrup = ['바닐라', '헤이즐넛', '카라멜', '클래식', '모카', '화이트 모카', '돌체 시럽']
food = food_df['이름'].to_list()
```

# 중요도 {beverage}{quantity}{size}{syrup}{food}

```python
import time
from tqdm import tqdm
result_li = []
for bev in beverage:
    for siz in size:
        for sy in syrup:
            for fo in food:    
                order = [f'{bev} {siz} 한잔 주세요',
                    f'{bev} {siz} 테이크 아웃이요',
                    f'{bev} {siz} 포장이요',
                    f'텀블러에 {bev} 한잔 주세요',
                    f'{bev} {siz} 하나, {bev} {siz} 하나요',
                    f'{bev} 5잔이요. 캐리어에 담아주세요.',
                    f'머그잔에 {bev} {siz} 주세요.',
                    f'{fo} 하나랑요, {bev} 하나, {bev} 한 잔이요.',
                    f'{fo}랑, {bev} 하나, {bev} 하나 주세요.',
                    f'{bev} {siz} 한 잔이요. {sy} 5번 넣어주세요. 텀블러에 담아갈게요.',
                    f'{bev} 한 잔이요. {siz}로 주세요. 테이크아웃잔에 주세요.',
                    f'{bev} {siz} 주세요. 자바칩 추가하는데 반은 갈아주시고 반은 위에 올려주세요. 휘핑은 빼주세요.',
                    f'{bev} {siz} 한 잔에 샷 추가해주세요.',
                    f'{bev} {siz} 둘, {bev} {siz} 둘이요',
                    f'{siz} {bev} 하나요',
                    f'{bev} {siz} 하나요. {sy} 좀만 넣어주세요.',
                    f'{bev} 한 잔이랑요, {fo}도 주세요. 사이즈는 {siz}요.']
                result_li.append(order)
```

```python
import itertools

wow = list(itertools.chain(*result_li))
```
