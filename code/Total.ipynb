```python
import os
result_dir = '/content/drive/MyDrive/Slot_tagging_project/data/result_data' 
train_dir = '/content/drive/MyDrive/Slot_tagging_project/data/datasets/train'
test_dir = '/content/drive/MyDrive/Slot_tagging_project/data/datasets/test'
val_dir = '/content/drive/MyDrive/Slot_tagging_project/data/datasets/val'

ind_df = os.path.join(test_dir,'seq.in')
os.path.join(test_dir,'seq.out')

li = [train_dir, test_dir, val_dir]
in_data = []
out_data = []

for i in li:
  with open(os.path.join(i,'data.txt'),'r') as f:
    print('data.txt : ',len(f.readlines()))

  with open(os.path.join(i,'seq.in'),'r') as f:
    print('seq.in : ',len(f.readlines()))
    for w in f.readlines():
      in_data.append(w.strip())
      

  with open(os.path.join(i,'seq.out'),'r') as f:
    print('seq.out',len(f.readlines()))
    for w in f.readlines():
      out_data.append(w.strip())

with open(os.path.join(result_dir,'seq.in'),'r') as f:
    print('seq.in : ',len(f.readlines()))

with open(os.path.join(result_dir,'seq.out'),'r') as f:
    print('seq.out',len(f.readlines()))



```
